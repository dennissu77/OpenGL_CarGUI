```python
import os
import json

path = '../json/result_vec_ori.json'
new_path = '../json/result_vec.json'
with open(path, 'r') as f:
    data = json.load(f)

for k, v in data.items():

    for obj_id, obj in v.items():
        data[k][obj_id]['x'] = data[k][obj_id]['x'] - 341
        data[k][obj_id]['y'] = data[k][obj_id]['y'] - 341

with open(new_path, 'w') as f:
    json.dump(data, f)
```

```python
import os
import numpy as np
from itertools import groupby

dir_path = '../data/my_surocc'
data_arr = []

for file in sorted(os.listdir(dir_path), key=lambda x: int(x.split('.')[0])):
    data = np.load(os.path.join(dir_path, file))
    data = data[(data[:, 0] % 2 == 0) & (data[:, 1] % 2 == 0) & (data[:, 1] > 50 )]
    
    data_sorted = data[data[:, -1].argsort()]
    data_sorted[:, 1], data_sorted[:, 2] = data_sorted[:, 2], data_sorted[:, 1].copy()
    data_sorted = np.array(data_sorted).tolist()
    grouped_data = {k: np.array(list(v))[:, :3].tolist() for k, v in groupby(data_sorted, key=lambda x: x[-1])}
    data_arr.append(grouped_data)
     
```
